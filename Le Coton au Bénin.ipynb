1. **Évolution de la Production Nationale de Coton** : J'examinerai comment la production nationale de coton a évolué entre **1957 et 2021**.

3. **Rendement Moyen du Coton par Hectare** : J'analyserai comment le rendement moyen du coton par hectare a changé au cours de la période **1957-2021**.

4. **Impact des Politiques Agricoles** : J'évaluerai l'impact des différentes politiques agricoles sur l'évolution de la culture du coton.

En répondant à ces questions, mon objectif est de mieux comprendre l'industrie cotonnière au Bénin et comment elle a évolué au fil du temps.

## Impact des différentes Politiques Agricoles : Quelle est l’impact des différentes politiques agricoles sur l’évolution de la culture du coton ?

```python
#Définissons la période des différents régimes.

Pr_Nicéphore_SOGLO = df.loc[1991 : 1995]

Pr_Mathieu_KEREKOU = df.loc[1996 : 2005]

Pr_Thomas_BONI_YAYI = df.loc[2006 : 2015]

Pr_Patrice_TALON = df.loc[2016 : ]
```

```python
#Déterminons les rendements moyens ; la superficie moyenne , la production moyenne et la production totale sur chacune des périodes

stats_Pr_Nicéphore_SOGLO = Pr_Nicéphore_SOGLO[["REND (Kg/Ha)", "SUP (Ha)", "PROD (T)"]].mean()
stats_Pr_Nicéphore_SOGLO["PROD TOTAL (T)"] = Pr_Nicéphore_SOGLO["PROD (T)"].sum()

stats_Pr_Mathieu_KEREKOU = Pr_Mathieu_KEREKOU[["REND (Kg/Ha)", "SUP (Ha)", "PROD (T)"]].mean()
stats_Pr_Mathieu_KEREKOU["PROD TOTAL (T)"] = Pr_Mathieu_KEREKOU["PROD (T)"].sum()

stats_Pr_Thomas_BONI_YAYI = Pr_Thomas_BONI_YAYI[["REND (Kg/Ha)", "SUP (Ha)", "PROD (T)"]].mean()
stats_Pr_Thomas_BONI_YAYI["PROD TOTAL (T)"] = Pr_Thomas_BONI_YAYI["PROD (T)"].sum()

stats_Pr_Patrice_TALON = Pr_Patrice_TALON[["REND (Kg/Ha)", "SUP (Ha)", "PROD (T)"]].mean()
stats_Pr_Patrice_TALON["PROD TOTAL (T)"] = Pr_Patrice_TALON["PROD (T)"].sum()
```

```python
# Affichons les statistiques
print("Statistiques pour la période Nicéphore SOGLO :\n", stats_Pr_Nicéphore_SOGLO)
print("Statistiques pour la période Mathieu KEREKOU :\n", stats_Pr_Mathieu_KEREKOU)
print("Statistiques pour la période Thomas BONI YAYI :\n", stats_Pr_Thomas_BONI_YAYI)
print("Statistiques pour la période Patrice TALON  :\n", stats_Pr_Patrice_TALON)
```

```python
# Création un DataFrame avec les statistiques pour chaque régime
data = {
    'Régime': ['Nicéphore SOGLO', 'Mathieu KEREKOU' , 'Yayi BONI', 'Patrice TALON'],
    'Superficie moyenne (Ha)': [stats_Pr_Nicéphore_SOGLO['SUP (Ha)'], stats_Pr_Mathieu_KEREKOU['SUP (Ha)'], stats_Pr_Thomas_BONI_YAYI['SUP (Ha)'], stats_Pr_Patrice_TALON['SUP (Ha)']],
    'Production moyenne (T)': [stats_Pr_Nicéphore_SOGLO['PROD (T)'], stats_Pr_Mathieu_KEREKOU['PROD (T)'], stats_Pr_Thomas_BONI_YAYI['PROD (T)'], stats_Pr_Patrice_TALON['PROD (T)']],
    'Production totale (T)': [stats_Pr_Nicéphore_SOGLO['PROD TOTAL (T)'], stats_Pr_Mathieu_KEREKOU['PROD TOTAL (T)'], stats_Pr_Thomas_BONI_YAYI['PROD TOTAL (T)'], stats_Pr_Patrice_TALON['PROD TOTAL (T)']]
}
df_stats = pd.DataFrame(data)
```

```python
# Faisons fondre le DataFrame pour le rendre compatible avec seaborn
df_melted = df_stats.melt(id_vars='Régime', var_name='Mesure', value_name='Valeur')
```

```python
# Créeons le graphique
sns.set_style("whitegrid") 
plt.figure(figsize=(8, 5))
sns.barplot(data=df_melted, x='Régime', y='Valeur', hue='Mesure', palette='viridis')
plt.title('Impact de la politique de chaque président sur la production du coton')
plt.text(x = 1, y = 0.5, s = 'Made by\nGenséric AGUIDISSOU', ha = 'right', va='center', rotation='vertical', transform=plt.gcf().transFigure, fontsize = 5)
plt.show()
```
